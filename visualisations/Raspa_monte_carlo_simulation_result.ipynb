# Temperature Variations at 10 KPa pressure(constant)

```python
df = pd.read_csv("temp_var_1e4.csv")
```

```python
df
```

```python
df.columns
```

```python
lis = ['Number of Molecules of methane( Average loading absolute)',
       'Molecules/unit cell', 'Total energy', 'Adsorbate-Adsorbate energy',
       'Host- Adsorbate']

for i in lis:
    fig, ax = plt.subplots(figsize=(9,7)) 
    sns.lineplot(df['Temperature'], df[i])
    _ = plt.xticks(fontsize=9)
    _ = plt.yticks(fontsize=9)
    plt.title(i + " vs Temp", fontsize=15)
```

```python
fig, ax = plt.subplots(figsize=(10,7)) 
sns.heatmap(df.corr(),annot=True, linewidths=2,cmap="coolwarm")
plt.title('Inter-Correlation matrix for Temp variations', fontsize=20)
_ = plt.xticks(fontsize=9)
_ = plt.yticks(fontsize=9)
```

# Pressure Variations at 300K Temperature(constant)
